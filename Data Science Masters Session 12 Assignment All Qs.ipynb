```python
## Data Science Masters Session 12 Assignment 

import pandas as pd
import sqlite3
dataSet = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data",header=None)
type(dataSet)
```

```python
dataSet.columns = ['age', 'workclass','fnlwgt','education','education_num','marital_status','occupation','relationship'
              ,'race','sex','capital_gain','capital_loss','hours_per_week','native_country','salary']
dataSet.head()
```

```python
import sqlite3
connection = sqlite3.connect('sqladb')
cursor = connection.cursor()
```

```python
dataSet.to_sql('adultData',connection)
```

```python
## Session 12 Assignment Q1. Select 10 records from the adult sqladb
q = """SELECT * FROM adultData LIMIT 10;"""
pd.read_sql(q, connection)
```

```python
## Session 12 Assignment Q2. Show me the average hours per week of all men who are working in private sector
q = """SELECT avg(hours_per_week) Average FROM adultData where sex='Male' and workclass='Private'"""
pd.read_sql(q, connection)
```

```python
## Session 12 Assignment Q3. Show me the frequency table for education, occupation and relationship, separately

q = """
        create table frequency(
        education varchar(50), 
        occupation varchar(50),
        relationship varchar(50)
        );
    """
pd.read_sql(q, connection)
#q="""select * from frequency"""
q1 = """
        insert into frequency(education,occupation,relationship)
        select education,occupation,relationship from adultData;
    """
pd.read_sql(q1, connection)
q3="""select * from frequency"""
#q="""drop table frequency"""
pd.read_sql(q3, connection)
```

```python
## Session 12 Assignment Q4.Are there any people who are married, working in private sector and having a masters degree

q="""
    select count(*) Result from adultData 
    where 
    (marital_status <>'Never-married' or marital_status <> 'Divorced') and workclass='Private' and education=' Masters'
    """
pd.read_sql(q, connection)
```

```python
## Session 12 Assignment Q5.What is the average, minimum and maximum age group for people working in different sectors

q="""select workclass,avg(age) Avg,min(age) Min,max(age) max from adultData group by workclass"""
#q="""select count(*) from adultData where workclass='?'"""
pd.read_sql(q, connection)
```

```python
## Session 12 Assignment Q6.Calculate age distribution by country

q="""select native_country,avg(age) Avg,min(age) Min, max(age) min from adultData group by native_country"""
pd.read_sql(q, connection)
```

```python
## Session 12 Assignment Q7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'


q="""select capital_gain,capital_loss,(capital_gain+' '+capital_loss) Net_capital_Gain from adultData"""
pd.read_sql(q, connection)
```
